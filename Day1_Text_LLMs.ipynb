```python
# Install required libraries
!pip install transformers sentence-transformers --quiet

from transformers import pipeline, MarianMTModel, MarianTokenizer
from sentence_transformers import SentenceTransformer, util

# TEXT GENERATION
generator = pipeline("text-generation", model="gpt2")
print("Generated Text:", generator("In the future, AI will", max_length=30)[0]['generated_text'])

# TRANSLATION
model_name = 'Helsinki-NLP/opus-mt-en-de'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

text = "Hello, how are you?"
tokens = tokenizer.prepare_seq2seq_batch([text], return_tensors="pt")
translated = model.generate(**tokens)
translated_text = tokenizer.batch_decode(translated, skip_special_tokens=True)
print("Translated to German:", translated_text[0])

# SENTENCE SIMILARITY
model = SentenceTransformer('all-MiniLM-L6-v2')
sentence1 = "AI is transforming the world."
sentence2 = "Artificial intelligence is changing everything."
embeddings1 = model.encode(sentence1, convert_to_tensor=True)
embeddings2 = model.encode(sentence2, convert_to_tensor=True)
similarity = util.pytorch_cos_sim(embeddings1, embeddings2)
print("Sentence similarity:", similarity.item())
```
